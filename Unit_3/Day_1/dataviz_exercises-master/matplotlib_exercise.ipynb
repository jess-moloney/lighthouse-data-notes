## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
df
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
x = df['month_number']
y = df['total_profit']

plt.title('Total Profit by Month')

plt.xlabel('Month')
plt.ylabel('Total Profit')

plt.plot(x,y)

plt.show()
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
x = df['month_number']
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

plt.plot(x,facecream, label='facecream')
plt.plot(x,facewash, label='facewash')
plt.plot(x,toothpaste, label='toothpaste')
plt.plot(x,bathingsoap, label='bathingsoap')
plt.plot(x,shampoo, label='shampoo')
plt.plot(x,moisturizer, label='moisturizer')

plt.legend(loc='upper left')

plt.title('Sales data by month for all products')

plt.xlabel('month')
plt.ylabel('sales')

# #get handles and labels
handles, labels = plt.gca().get_legend_handles_labels()

# #specify order of items in legend
order = [3,2,0,4,5]

# #add legend to plot
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order]) 
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
x = df['total_profit']

plt.hist(x, bins = 10)

plt.title('Frequency Distribution of Monthly Profits')

plt.xlabel('Monthly Profit')
plt.ylabel('Frequency')

plt.show()
```
