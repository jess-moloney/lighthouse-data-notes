## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data/company_sales_data.csv')
```

```python
df.head()
```

### Task 1 (Functional Approach): Read the `total_profit` of all months and display it using a line plot.

```python
x = df['month_number']
y = df['total_profit']

plt.title('Total Profit by Month')

plt.xlabel('Month')
plt.ylabel('Total Profit')

plt.plot(x,y)

plt.show()
```

### Task 1 (Stateless Approach): Read the `total_profit` of all months and display it using a line plot.

```python
# get the data

x = df['month_number']
y = df['total_profit']

# create the figure and axes, set the figure size

fig, ax = plt.subplots(figsize=(10,6))

# set the title, axis labels

ax.set_title('Total Profit by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Total Profit')

# plot the line

ax.plot(x,y)
plt.show()
```

### Task 2 (Functional): Read all the different product sales data and display it using a multiline plot.

```python
df['facewash']
```

```python
x = df['month_number']
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

plt.plot(x,facecream, label='facecream')
plt.plot(x,facewash, label='facewash')
plt.plot(x,toothpaste, label='toothpaste')
plt.plot(x,bathingsoap, label='bathingsoap')
plt.plot(x,shampoo, label='shampoo')
plt.plot(x,moisturizer, label='moisturizer')

plt.legend(loc='upper left')

plt.title('Sales data by month for all products')

plt.xlabel('month')
plt.ylabel('sales')

# #get handles and labels
handles, labels = plt.gca().get_legend_handles_labels()

# #specify order of items in legend
order = [3,2,0,4,5]

# #add legend to plot
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order]) 
```

### Task 2 (Stateless): Read all the different product sales data and display it using a multiline plot.

```python
# get the data
x = df['month_number']
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# multiline plot
ax.plot(x,bathingsoap, label = 'bathingsoap')
ax.plot(x,toothpaste, label = 'toothpaste')
ax.plot(x,facecream, label='facecream')
ax.plot(x,shampoo, label = 'shampoo')
ax.plot(x,moisturizer, label = 'moisturizer')
ax.plot(x,facewash, label = 'facewash')

# titles
ax.set_title('Product Sales by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Product Sales')

# legend
ax.legend()
```

### Task 3 (Functional): Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
x = df['total_profit']

plt.hist(x, bins = 10)

plt.title('Frequency Distribution of Monthly Profits')

plt.xlabel('Monthly Profit')
plt.ylabel('Frequency')

plt.show()
```

### Task 3 (Stateless): Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
import numpy as np

#get data
y = df['total_profit']

#create the figure and axes
fig, ax = plt.subplots(figsize=(10,6))

#plot the histogram
ax.hist(y,bins=20)

#set the x axis tick labels
xticklabels = np.arange(175000, 425000, 25000)
ax.xaxis.set_ticks(xticklabels)
plt.show()
```
