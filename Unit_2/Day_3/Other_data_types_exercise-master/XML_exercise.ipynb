# XML exercise

Using data from [**mondial database**](https://drive.google.com/file/d/14lFT4nWHgwN36ij4XZh6OUuup-K9qLgR/view?usp=sharing) find the answers to following questions:

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

# QUESTION 1

10 countries with the lowest infant mortality rates

```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
tree = ET.parse("mondial.xml")
tree
```

```python
root = tree.getroot()
root
```

```python
mondial_dict = {'country': [],
                'infant_mortality': []}

for country in root:
    mondial_dict['country'].append(country.find('name').text)
    
    infant_mortality = country.find('infant_mortality')
    if infant_mortality is None:
        mondial_dict['infant_mortality'].append(None)
    else:
        mondial_dict['infant_mortality'].append(float(infant_mortality.text))

mond_df = pd.DataFrame(mondial_dict)
mond_df
```

```python
mond_df.nsmallest(10, 'infant_mortality', keep='all')
```

# QUESTION 2

10 cities with the largest population

```python
mondial_dict = {'name': [],
                'population': []}

for country in root:
    for province in country.findall('province'):
        if province is None:
            continue
        else:
            for city in province.findall('city'):
                name = city.find('name')
                for element in city.findall('population'):
                    if element.attrib['year'][-1]:
                        population = element

                if name is None:
                    mondial_dict['name'].append(None)
                else:
                    mondial_dict['name'].append(name.text)

                if population is None:
                    mondial_dict['population'].append(None)
                else:
                    mondial_dict['population'].append(int(population.text))
                
    for city in country.findall('city'):
        name = city.find('name')
        for element in city.findall('population'):
            if element.attrib['year'][-1]:
                population = element
                
        if name is None:
            mondial_dict['name'].append(None)
        else:
            mondial_dict['name'].append(name.text)
            
        if population is None:
            mondial_dict['population'].append(None)
        else:
            mondial_dict['population'].append(int(population.text))


mond_df = pd.DataFrame(mondial_dict)
mond_df.nlargest(10, 'population', keep='all')
```
