```python
import pandas as pd
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('release_dates.csv', index_col=None,
                                      parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

### Count the number of movies with "Christmas" in their title for each month, that is released are the USA.

```python
release_dates['year'] = release_dates['date'].dt.year
```

```python
release_dates['month'] = release_dates['date'].dt.month
```

```python
release_dates['day_name'] = release_dates['date'].dt.day_name()
```

```python
release_dates['day'] = release_dates['date'].dt.day_of_week
```

```python
release_dates.head()
```

```python
filter_df = release_dates[(release_dates['title'].str.contains('Christmas')) & (release_dates['country'] == 'USA')]
filter_df.groupby('month').agg(n_movies = ('title','count'))
```

### Count the number of movies with "Christmas" in their title for each month, that are released in Canada.

```python
filter_df = release_dates[(release_dates['title'].str.contains('Christmas')) & (release_dates['country'] == 'Canada')]
filter_df.groupby('month').agg(n_movies = ('title','count'))
```

### Count the number of movies in which the titles start with "The Hobbit" for each month, that are released in the USA.

```python
filter_df = release_dates[(release_dates['title'].str.contains('The Hobbit')) & (release_dates['country'] == 'USA')]
filter_df.groupby('month').agg(n_movies = ('title','count'))
```

```python
release_dates[(release_dates['title'].str.startswith('The Hobbit')) & (release_dates['country'] == 'USA')]
```

### Count the number of movies with "Romance" in their title for **each day of the week**, that are released in the USA.

```python
filter_df = release_dates[(release_dates['title'].str.contains('Romance')) & (release_dates['country'] == 'USA')]
filter_df.groupby(['day_name','day']).agg(n_movies = ('title','count')).sort_values('day')
```

### Count the number of movies with "Action" in their title for **each day of the week**, that are released in the USA.

```python
filter_df = release_dates[(release_dates['title'].str.contains('Action')) & (release_dates['country'] == 'USA')]
filter_df.groupby(['day_name','day']).agg(n_movies = ('title','count')).sort_values('day')
```

### In which months do films with the actress Judi Dench tend to be released in the USA?

```python
# filter the cast table to get the films with Judi Dench
filter_cast = cast[(cast['name'] == 'Judi Dench')]
```

```python
# create a merged table showing the release dates and countries for each Judi Dench film
merged = pd.merge(filter_cast, release_dates, on=('title','year'))
```

```python
merged.head()
```

```python
# filter the merged table to show only USA films
merged_filter = merged[merged['country']=='USA']
```

```python
merged_filter.sort_values('month')
```

```python
merged_filter.groupby('month').agg(n_movies = ('title','count'))
```
