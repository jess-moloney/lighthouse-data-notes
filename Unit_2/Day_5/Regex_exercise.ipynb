## Print all the words

Words with either 'Bar' or 'Baz' in them

```python
word_list =[]

for line in lines:
   word_list.append(line.split())

word_list
```

```python
regex = r"Bar|Baz"

for lst in word_list:
    for word in lst:
        if re.search(regex, word):
            print(word)
```

Words with either 'Whe' or 'The' in them

```python
regex = r"Whe|The"

for lst in word_list:
    for word in lst:
        if re.search(regex, word):
            print(word)
```

Words containing a double character (e.g. 'oo')

```python
regex = r"(\w)\1"

for lst in word_list:
    for word in lst:
        if re.search(regex, word):
            print(word)
```

## Cleanup string codes so they contain only numbers
* Remove slash and spaces

```python
codes = ['2373/ 8293', 
         ' 8292342 / 8263', 
         '12/903820 ',
         '8203184 / 02342 ']

regex = r'[// ]{1,}'
for code in codes:
    print('original:',code,'new:',re.sub(regex,'',code))
```

## Switch order of the numbers in string
* Preserve slash /
* Remove spaces

```python
codes = ['2373/ 8293', 
         ' 8292342 / 8263', 
         '12/903820 ',
         '8203184 / 02342 ']
regex = r'(\d+)\s?/\s?(\d+)'

for code in codes:
    print('original:',code, 'new: ',re.sub(regex, r'\2/\1',code))
```
